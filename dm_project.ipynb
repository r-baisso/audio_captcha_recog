```python
import matplotlib.pylab as plt
import librosa
import pandas as pd
import math


data, fs = librosa.load('6.wav', None)

print(f"frequencia de amostragem = {fs}")
#neste caso o audio eh mono e por isso apenas utilizamos a dimensao 0
print(f"quantidade de amostras = {data.shape[0]}")

#descobrindo a duracao total do audio em segundos (amostras/amostras/segundos = segundos)
duracao_total = data.shape[0]/fs
#print(duracao_total)

intervalo = 1
dados_p_seg = {}
#iterando sobre um vetor que vai de 0 ao tamanho do
for i, ini in enumerate(range(0, data.shape[0], fs*intervalo)):
    #print(f"{ini}:{ini + fs*intervalo}")
    dados_p_seg [i] = pd.Series(data[ini : (ini + fs*intervalo)])


#dados_p_seg[0].plot()
#dados_p_seg[1].plot()
#dados_p_seg[2].plot()
df = pd.Series(data[0 : 95232])
df.plot()
plt.show()

#adquirindo apeas parte positiva do sinal
df = df.map(lambda a: abs(a))
mean = df.mean()
print(mean)
df.plot()
plt.show()
#print(dados_p_seg)

#diminuindo no segmento
df = df.map(lambda a: a if a > mean else 0)
df.plot()
plt.show()
```
